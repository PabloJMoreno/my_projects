```python
from river import imblearn, linear_model
```

Usando **undersampling** de la clase mayoritaria para ajustar 80/20 el dataset

```python
# Usando un modelo de clasificaciÃ³n 'Reg LogÃ­stica' con undersampling
model_u = (
    preprocessing.StandardScaler() |
    imblearn.RandomUnderSampler(
        classifier = linear_model.LogisticRegression(),
        desired_dist = {0: .8, 1: .2},
        seed = 42
    )
)
```

```python
metric = metrics.ROCAUC()
```

```python
%%time 

for xi, yi in stream.iter_pandas(x, y, seed = 100):
    yi_pred = model_u.predict_proba_one(xi) # se itera sobre cada observaciÃ³n 
    model_u.learn_one(xi, yi)               # el modelo se ajusta por cada observaciÃ³n iterada
    metric.update(yi, yi_pred)              # el modelo se actualiza creando 'memoria'
    
# metrica de valoracion
metric
```

```python
# evaluaciÃ³n global del modelo 'undersampling' de manera progresiva (emulando en producciÃ³n)
evaluate.progressive_val_score(stream.iter_pandas(x, y), model_u, metric) 
```

```python
# evaluaciÃ³n progresiva del modelo a medida que entran nuevos datos

steps = evaluate.iter_progressive_val_score(
    model = model_u,
    dataset = stream.iter_pandas(x, y),
    metric = metric,
    step = 10000      # numero de iteraciones para arrojar el resultado
)

for step in steps:
    print(step)
```

Usando **oversampling** de la clase minoritaria -creando datos sintÃ©ticos-

```python
# Usando un modelo de clasificaciÃ³n 'Reg LogÃ­stica' con oversampling
model_o = (
    preprocessing.StandardScaler() |
    imblearn.RandomOverSampler(
        classifier = linear_model.LogisticRegression(),
        desired_dist = {0: .8, 1: .2},
        seed = 42
    )
)
```

```python
%%time 

for xi, yi in stream.iter_pandas(x, y, seed = 100):
    yi_pred = model_o.predict_proba_one(xi) # se itera sobre cada observaciÃ³n 
    model_o.learn_one(xi, yi)               # el modelo se ajusta por cada observaciÃ³n iterada
    metric.update(yi, yi_pred)              # el modelo se actualiza creando 'memoria'
    
# metrica de valoracion
metric
```

```python
# evaluaciÃ³n global del modelo 'oversampling' de manera progresiva (emulando en producciÃ³n)
evaluate.progressive_val_score(stream.iter_pandas(x, y), model_o, metric) 
```

```python
# evaluaciÃ³n progresiva del modelo a medida que entran nuevos datos

steps = evaluate.iter_progressive_val_score(
    model = model_o,
    dataset = stream.iter_pandas(x, y),
    metric = metric,
    step = 10000      # numero de iteraciones para arrojar el resultado
)

for step in steps:
    print(step)
```

Enfoque hÃ­brido combinando **undersampling** de la categoria mayoritaria y **oversampling** de la minoritaria

```python
from river import optim
```

```python
model_h = (
    preprocessing.StandardScaler() |
    imblearn.RandomUnderSampler(                       # usar undersampler para la categoria mayoritaria
        classifier = linear_model.LogisticRegression(
            loss = optim.losses.Log(weight_pos=5)      # el parÃ¡metro 'weight_pos' significa que x5 la clase positiva (oversampling)
        ),
        desired_dist={0: .8, 1: .2},
        seed=42
    )
)
```

```python
%%time

for xi, yi in stream.iter_pandas(x, y, seed = 100):
    yi_pred = model_h.predict_proba_one(xi) # se itera sobre cada observaciÃ³n 
    model_h.learn_one(xi, yi)               # el modelo se ajusta por cada observaciÃ³n iterada
    metric.update(yi, yi_pred)              # el modelo se actualiza creando 'memoria'
    
# metrica de valoracion
metric
```

```python
# evaluaciÃ³n global del modelo 'hibrido' de manera progresiva (emulando en producciÃ³n)
evaluate.progressive_val_score(stream.iter_pandas(x, y), model_h, metric) 
```

```python
# evaluaciÃ³n progresiva del modelo a medida que entran nuevos datos

steps = evaluate.iter_progressive_val_score(
    model = model_h,
    dataset = stream.iter_pandas(x, y),
    metric = metric,
    step = 10000      # numero de iteraciones para arrojar el resultado
)

for step in steps:
    print(step)
```
