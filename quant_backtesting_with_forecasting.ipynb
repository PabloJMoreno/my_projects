```python
# importar funciones de backtesting y crossover

from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA, GOOG
```

```python
GOOG
```

```python
GOOG.info()
```

```python
# crear UDF de backtesting con cruce de SMA

class SmaCross(Strategy):
    def init(self):
        price = self.data.Close
        self.ma1 = self.I(SMA, price, 10)
        self.ma2 = self.I(SMA, price, 20)

    def next(self):
        if crossover(self.ma1, self.ma2):
            self.buy()
        elif crossover(self.ma2, self.ma1):
            self.sell()
```

```python
# backtesting de Google con commission de 0.002
bt = Backtest(GOOG, SmaCross, commission=.002, exclusive_orders=True)
```

```python
# instanciamiento de estadísticas
stats = bt.run()
```

```python
#visualización de estadísticas para Google
stats
```

```python
# visualización del analisis de backtesting con indicadores long / short
bt.plot()
```

```python
from prophet import Prophet
```

```python
import numpy as np
import pandas as pd
```

```python
msft_ts = msft_hist[['Close']]
```

```python
msft_ts['ds'] = msft_ts.index
```

```python
msft_ts.reset_index(drop=True, inplace=True)
```

```python
msft_ts.rename(columns={'Close':'y'}, inplace=True)
```

```python
msft_ts.info()
```

```python
# reformat timestamp
msft_ts['ds'] = pd.to_datetime(msft_ts['ds'].dt.strftime('%Y-%m-%d'))
```

```python
# remove time zone
msft_ts['ds'] = msft_ts['ds'].dt.tz_localize(None)
```

```python
msft_ts
```

```python
msft_ts.info()
```

```python
m = Prophet()
```

```python
m.fit(msft_ts)
```

```python
future = m.make_future_dataframe(periods = 20, freq = 'B', include_history=True)
```

```python
msft_forecast = m.predict(future)
```

```python
msft_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(20)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
plot_plotly(m, msft_forecast)
```
