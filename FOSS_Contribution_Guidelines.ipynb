# **1 Descripción general**

## 1.1 Objetivo

La Compañía apoya ciertas actividades asociadas con la contribución y publicación de Software Libre y de Código Abierto (FOSS), cuando hacerlo suponga un beneficio para la Compañía y no presente un riesgo inaceptable. Los riesgos incluyen la divulgación involuntaria de información confidencial, la renuncia a los derechos exclusivos de propiedad intelectual, la infracción de los derechos de terceros y el daño reputacional.

Estos lineamientos establecen unas actividades mínimas, aprobaciones, restricciones y evidencias de cumplimiento que deben cubrir los procesos de gobierno asociados.

# **2 Gobernanza**

## 2.1 Oficina del programa de código abierto (OSPO)

Los procedimientos y procesos de gobierno descritos en estas pautas deben ser supervisados p​or el Presidente de la Oficina del Programa de Código Abierto (OSPO).

La Oficina del Programa de Código Abierto (OSPO) es responsable de establecer, desarrollar y ejecutar la estrategia de contribución y el gobierno de Software Libre y de Código Abierto (FOSS) en la Compañía bajo la dirección del propietario de las pautas.

## 2.2 Riesgo de contribución de FOSS

Contribuir a proyectos FOSS presenta varios riesgos. Se deben implementar controles internos preventivos y de detección para abordar los riesgos que se aplican a estas actividades. Además, se deben identificar y registrar los riesgos que son específicos de la participación en un proyecto FOSS en particular, así como implementar controles internos preventivos y de detección, tal como se define en estas pautas.

A título orientativo, los principales riesgos son:

1. **Fuga de datos** : esto incluye la divulgación no autorizada de información no pública y datos de clientes, la divulgación no autorizada de procedimientos de seguridad internos, información interna como estructuras de red, direcciones IP, nombres de servidores, material clave, datos de autenticación, etc.

2. **Pérdida financiera** : la contribución de software a proyectos FOSS bajo las licencias FOSS más aceptables implica otorgar licencias perpetuas y libres de regalías a los consumidores del proyecto FOSS y, por lo tanto, se debe tener en cuenta cualquier amortización financiera del activo de software.

3. **Propiedad intelectual (PI)** : se debe considerar antes de contribuir con obras a los proyectos de FOSS si la empresa debe conservar los derechos de propiedad intelectual sobre cualquier parte de las obras, incluso si se deben solicitar las marcas comerciales y/o las patentes por adelantado.

4. **Daño a la reputación** : existen varios riesgos que podrían reflejarse negativamente en la marca y la reputación de la Compañía, que van desde la decisión de colaborar en un proyecto, la calidad del código aportado, comentarios, opiniones y otras representaciones atribuidas a la Compañía, los empleados y los proyectos que se abandonan por falta de inversión.

5. **Riesgo de conflicto de intereses** : riesgo de que los empleados de la Compañía participen en actividades o intereses comerciales fuera del alcance de sus actividades de contribución permitidas de FOSS que interfieran o comprometan, o parezcan interferir o comprometer, su capacidad de actuar en el mejor interés de los empresa y/o sus clientes.

Además, existen varios riesgos consecuenciales a tener en cuenta que implican uno más de los riesgos principales:

1. **Cumplimiento de licencias FOSS** : el incumplimiento de las licencias FOSS puede exponer a la empresa a varios riesgos, incluidos litigios, daños a la reputación y pérdidas financieras.

2. **Permisos de licencia, garantía y responsabilidad de FOSS** : hay una gran cantidad de licencias de código abierto, cada una con diferentes permisos, restricciones y condiciones que pueden proteger o exponer a la empresa a litigios o restringir nuestra capacidad para reclamar derechos de patente o marca comercial.

3. **Incumplimiento de políticas relevantes de la empresa** : se analizan en Alineación específica con las reglamentaciones pertinentes e incluyen compromisos reglamentarios, como controles de exportación y comercio, y flujos de datos transfronterizos.

# **3 Requisitos estándar**

La Compañía apoya la contribución y publicación de software libre y de código abierto (FOSS) dentro de los límites de estas pautas.

## 3.1 Terminología

Los términos FOSS pueden entrar en conflicto con las definiciones establecidas, por ejemplo: Iniciativa , Programa y Proyecto son términos bien establecidos de gestión de programas y proyectos. Para mayor claridad, los términos utilizados en estas pautas se referirán a FOSS.

## 3.2 Taxonomía - Contribución de código abierto

La Compañía ha establecido una taxonomía para ser utilizada junto con estas pautas y procedimientos asociados.

### Categorías de proyectos de FOSS


Las contribuciones realizadas a los proyectos de FOSS por parte de los empleados de la empresa deben pertenecer a una de las cuatro categorías a continuación y cumplir con los requisitos correspondientes para la categoría.

### a. Categoría	Descripción



empresa moderada	Proyectos de código abierto iniciados por la Compañía en los que la Compañía actuará como moderador/mantenedor principal del proyecto. Estas pueden ser soluciones de software completas, propuestas de estándares abiertos o funciones simples.

### b. Interés de la empresa (mayor)

Proyectos de código abierto consumidos en aplicaciones de la Empresa, moderados por una entidad ajena a la Empresa, donde:
1. La Empresa desea influir en la dirección del proyecto FOSS o,
2. La Empresa desea contribuir con una funcionalidad significativa que beneficiaría a La Empresa y al proyecto.

Estas pueden ser soluciones de software completas, propuestas de estándares abiertos o funciones simples.

### c. Interés de la empresa (menor)

Proyectos de código abierto consumidos en aplicaciones de la empresa, moderados por una entidad ajena a la empresa, donde:
1. La empresa no desea influir en la dirección del proyecto pero,
2. La empresa se beneficia al permitir que los empleados contribuyan.

Los ejemplos incluyen pequeñas mejoras y correcciones de errores en FOSS que se usan comúnmente en las aplicaciones de la empresa para evitar la necesidad de bifurcaciones internas para solucionar los defectos.

### d. Contribución personal

Un esfuerzo de trabajo de código abierto realizado por empleados de la empresa fuera de sus actividades regulares contratadas por la empresa para proyectos de FOSS que no están categorizados como moderados por la empresa o de interés de la empresa y para los cuales no hay ningún beneficio para la empresa al contribuir.

## 3.3 Infraestructura de la empresa

1. Las actividades de contribución de FOSS moderadas por la empresa y de interés de la empresa se pueden realizar dentro o fuera de la red de la empresa.
2. Las contribuciones personales de FOSS se deben realizar fuera de la red de la Compañía usando hardware que no sea administrado por la Compañía.
3. La infraestructura externa utilizada para albergar proyectos moderados por la empresa debe ser aprobada por la OSPO.

## 3.4 Herramientas estándar OSPO

1. Se deben implementar controles internos preventivos y de detección para abordar los riesgos descritos en Riesgo de contribución de FOSS.

2. Las actividades de contribución de FOSS moderadas por la empresa y de interés de la empresa deben usar herramientas estándar aprobadas por OSPO al realizar contribuciones dentro de la red de la empresa o en el hardware administrado por la empresa.

3. Estas pautas no imponen restricciones de herramientas cuando se realizan contribuciones fuera de la red de la Compañía en hardware que no es administrado por la Compañía y que no está definido en la Infraestructura de la Compañía.

4. Estas pautas no imponen restricciones de herramientas para las contribuciones personales que no sean las definidas en la infraestructura de la empresa.

5. Los proyectos FOSS moderados por la empresa deben utilizar una herramienta de gestión de código fuente aprobada por OSPO.

6. Los proyectos FOSS moderados por la empresa deben mantener un historial de transacciones que muestre evidencia de la revisión por pares, el aprobador y la fecha.

7. La comunidad de FOSS correspondiente define los requisitos de gestión del código fuente, revisión por pares y escaneo para los proyectos de interés de la empresa.

8. Todas las actividades de contribución que ocurran dentro de la red de la Compañía o en el hardware administrado por la Compañía deben estar sujetas a análisis y revisión de cumplimiento. Esto incluye procesos manuales o automatizados para validar:

   i. Cumplimiento de la licencia FOSS

   ii. Herramientas de creación/verificación (p. ej., el revisor por pares no es la misma persona que el colaborador)

   iii. Escaneos para reducir el riesgo de revelar información no pública de la empresa, violación de licencia de código abierto, texto de derechos de autor validado y cumplimiento de las normas de exportación.

## 3.5 Alineación específica con regulación pertinente

### Gestión de identidad y acceso

Los requisitos de administración de acceso e identidad no se aplican a las plataformas de herramientas alojadas externamente aprovechadas por los esfuerzos de contribución de FOSS. Las actividades de contribución y gobierno de FOSS internas de la empresa deben cumplir con los requisitos de gestión de acceso e identidad.

### Gestión de registros

La Compañía está obligada a conservar ciertos registros por motivos legales, reglamentarios o comerciales, previamente aprobados. Esto no se aplica a las actividades de contribución de FOSS y sí se aplica a las actividades de gobierno de OSPO.

### Regulaciones de Exportación/Importación de Tecnología

El desarrollo de código abierto es una actividad global que permite que el software esté disponible públicamente a través de las fronteras nacionales. Algunos países pueden requerir contribuciones de proyectos FOSS para cumplir con sus regulaciones locales de control de exportaciones. Esto se describe con más detalle en [Cumplimiento de las normas de exportación de tecnología](https://github-com.translate.goog/Citi/citi-ospo/blob/main/FOSS%20Contribution%20Guidelines.md?_x_tr_sl=en&_x_tr_tl=es&_x_tr_hl=en&_x_tr_pto=wapp#technologyexportregulationscompliance).

### Clasificación de la información

Las contribuciones de código abierto para las categorías de proyectos moderados por la empresa y de interés de la empresa deben revisarse y aprobarse para su publicación antes de transmitirse fuera de la red de la empresa:

1. Las contribuciones deben estar asociadas a un proyecto FOSS inventariado
 y registrado en el inventario OSPO FOSS.

2. Las contribuciones que contengan información material no pública, criptografía no estándar (como se define en Cumplimiento de las normas de exportación de tecnología ) o mecanismos de autenticación no estándar deben corregirse (p. ej., eliminar la información confidencial) antes de la aprobación.

### Flujos de datos transfronterizos


Muchos países imponen requisitos reglamentarios que rigen el flujo de datos personales fuera de su jurisdicción.

1. Estos requisitos no se aplican a las contribuciones de FOSS que incluyen solo las siguientes actividades:

i. Contribuir código fuente o documentación que ha sido clasificada por la Compañía como pública antes de la contribución,

ii. Participar en reuniones de grupos de trabajo con participantes que no pertenecen a la Compañía cuando el tema no incluye asuntos comerciales internos de la Compañía,

iii. El uso de datos clasificados por la Empresa como públicos.

2. Estos requisitos si se aplican a todas las demás Contribuciones de FOSS que no califican para las actividades enumeradas anteriormente, incluidas las actividades de Contribuciones de FOSS que contienen datos en claro o datos anónimos o no identificados. Estas actividades deben ser aprobadas antes de la contribución.

### Comunicaciones Corporativas

Se deben aprobar los compromisos de proyectos moderados por la empresa o de interés para la empresa que den lugar a comunicados de prensa u otros anuncios públicos.

### Comunicaciones Electrónicas

La Compañía tiene la obligación reglamentaria de conservar registros de conversaciones telefónicas y electrónicas que se relacionen con la recepción, transmisión y ejecución de órdenes de clientes así como transacciones por cuenta propia, incluidas las comunicaciones destinadas a generar una transacción comercial, incluso si en última instancia no lo hacen.

Las actividades de proyectos moderados por la compañía, de interés de la compañía y de contribución personal son de naturaleza general. Nunca deben discutir negocios de la Compañía y, por lo tanto, están fuera del alcance de estas regulaciones.

Las comunicaciones electrónicas realizadas dentro de la red de la Compañía en hardware administrado por la Compañía deben usar canales aprobados y nunca deben tratar asuntos comerciales de la Compañía.

Las comunicaciones electrónicas realizadas fuera de la red de la Compañía en hardware no administrado por la Compañía no están sujetas a estas restricciones, pero nunca deben hablar sobre negocios de la Compañía.

### Gestión de terceros

La relación con un tercero implica la provisión de servicios o productos -u otra relación comercial en curso- entre la Compañía y un tercero, como se indica en uno o más acuerdos legales que describen específicamente los productos, servicios y expectativas de desempeño de terceras partes.

1. El personal de la empresa que contribuye a las comunidades de código abierto se distingue del desempeño de terceros y, por lo tanto, las expectativas están fuera del alcance.

2. Los proyectos moderados por la empresa que se basan en repositorios y herramientas alojados externamente para respaldar las contribuciones están dentro del alcance.

## 3.6 Actividades de Gobernanza

**Los proyectos FOSS deben ser revisados ​​y aprobados por OSPO antes de que se puedan realizar contribuciones. Las contribuciones de la empresa están prohibidas cuando OSPO retira su aprobación o el proyecto FOSS se retira/archiva. La revisión de OSPO no se aplica a las contribuciones personales (ver [Categorías de proyectos de FOSS](https://github-com.translate.goog/Citi/citi-ospo/blob/main/FOSS%20Contribution%20Guidelines.md?_x_tr_sl=en&_x_tr_tl=es&_x_tr_hl=en&_x_tr_pto=wapp#fossprojectcategories)).**

### Gestión e Inventario de Licencias FOSS

Las licencias gratuitas y de código abierto son un acuerdo legal entre los autores y los consumidores que dictan los términos y condiciones que acompañan el uso, la modificación y la distribución del software.

1. OSPO debe mantener un inventario de las licencias FOSS aprobadas para la actividad de contribución por parte de los contribuyentes de la Compañía.

2. El inventario debe proporcionar un flujo de trabajo auditable de revisión legal y aprobación comercial o rechazo de licencias para actividades de contribución que detalle sus permisos, condiciones y limitaciones.

3. La revisión y aprobación o rechazo de licencias para actividades de contribución deben abordar los riesgos descritos en R[iesgo de contribución de FOSS](https://github-com.translate.goog/Citi/citi-ospo/blob/main/FOSS%20Contribution%20Guidelines.md?_x_tr_sl=en&_x_tr_tl=es&_x_tr_hl=en&_x_tr_pto=wapp#fosscontributionrisk).

### Proyectos (principales) moderados por la empresa y de interés de la empresa

**Evaluación y registro de oportunidades**

1. Todas las solicitudes para participar y contribuir a proyectos FOSS deben registrarse en una herramienta aprobada por OSPO.

2. OSPO debe revisar y aprobar las condiciones de membresía de la Fundación o comunidad. Esto incluye solicitar una revisión legal y aceptar cualquier riesgo identificado relacionado con los CLA/CCLA (si corresponde).

3. La solicitud debe incluir una revisión de la propiedad intelectual y los intereses en competencia, y determinar si existe la necesidad de una solicitud de patente o marca registrada antes de comenzar la contribución.

4. La solicitud deberá contar con la aprobación de un gestor con control presupuestario que pueda aceptar el saneamiento económico de las obras potencialmente aportadas.

5. Si la licencia del proyecto no se aprueba en el inventario, la OSPO debe buscar la revisión legal y la categorización de la licencia.

**Revisión y decisión de OSPO**

1. La OSPO revisará las solicitudes y devolverá una disposición basada en el asesoramiento de Legal y otras PYMES cuando sea necesario.

2. La OSPO tiene autoridad para tomar decisiones sobre las solicitudes y debe registrar su decisión en el sistema de registro.

**Iniciación**

Los proyectos FOSS pueden estar respaldados por una comunidad independiente o existir bajo una organización matriz (por ejemplo, la Fundación Linux).

1. Cuando el proyecto sea administrado por una organización matriz, la OSPO designará un Administrador de la Fundación para mantener la relación de la Compañía con la organización matriz en buen orden y de conformidad con los estatutos de la organización.

2. Una vez que se haya establecido la relación con la Fundación, el Administrador de la Fundación debe designar a un Administrador de proyectos FOSS para mantener en buen orden la relación de la Empresa con los proyectos. Esto incluye el cumplimiento de sus estatutos y requisitos de licencia.

3. Para los proyectos FOSS independientes que no se mantienen bajo una organización matriz, OSPO designará a un administrador que no sea de la fundación que supervisará la gobernanza de todos los proyectos de este tipo. El administrador que no sea de la fundación debe designar un administrador de proyectos FOSS para establecer y mantener en buen estado la relación de la empresa con los proyectos.

**Mantenimiento**

1. Los administradores de la fundación de FOSS, los administradores que no pertenecen a la fundación, los administradores de proyectos y el personal relacionado deben monitorear sus compromisos para detectar cambios de riesgo relevantes (p. ej., proyectos inactivos, cambios de licencia, impactos transfronterizos, cambios en las regulaciones de exportación, etc.).

2. Los posibles cambios en el riesgo deben capturarse en el inventario de FOSS al momento de la identificación, y el administrador del proyecto responsable debe involucrar a las PYME relevantes (por ejemplo, legal, transfronteriza, etc.) para una evaluación de impacto del cambio.

3. Se debe realizar una revisión de OSPO cuando se ha capturado un cambio potencial en el riesgo en el inventario FOSS. La disposición y las acciones tomadas deben registrarse en el inventario de FOSS.

4. La OSPO debe asegurarse de que todos los compromisos de FOSS sean revisados ​​por los Administradores de la Fundación FOSS designados, los Administradores que no pertenecen a la Fundación o los Administradores de Proyectos al menos una vez al año para identificar cambios de riesgo sustanciales en el estado del compromiso de FOSS. La disposición y las acciones tomadas deben registrarse en el Inventario de FOSS.

5. Los administradores de la fundación deben preparar solicitudes de jubilación/migración para los proyectos FOSS sobre los que se pretende hacer transición. Estos deben ser revisados ​​por la OSPO, y la disposición y las acciones tomadas deben registrarse en el Inventario de FOSS.

**Transición/Retiro**

1. La OSPO debe revisar y aprobar la decisión de retirarse de un proyecto FOSS, una fundación FOSS o una comunidad de código abierto.

2. Se debe acordar una fecha de vigencia y registrarla en el inventario FOSS después de haber sido solicitada por la parte interesada.

3. Al determinar que la empresa ya no tiene interés en un proyecto moderado por ésta, el proyecto debe transferirse a una comunidad que no sea de la empresa para su mantenimiento continuo o archivarse. La realización de estas actividades debe evitar el riesgo de marca y reputación de la Compañía.

4. Los administradores de fundaciones y proyectos FOSS retirados deben asegurarse de que se eliminen los derechos de las herramientas aprobadas por OSPO para garantizar que la contribución no se puedan realizar utilizando la infraestructura de la empresa en la fecha de vigencia acordada.

**Cerrado**

El compromiso o proyecto de código abierto se debe realizar con éxito en la transición o retirarse en la fecha de vigencia acordada y el cierre registrado en el inventario de FOSS.

### Proyectos (de menor) interés para la empresa

**Evaluación y registro de oportunidades**

1. Las solicitudes de acceso de cualquier empleado de la empresa para contribuir a un proyecto FOSS de interés (menor) de la empresa deben enviarse y registrarse mediante una herramienta aprobada por OSPO.

2. Las contribuciones a los intereses (menores) de la empresa no requieren la aprobación explícita de OSPO si el producto FOSS está aprobado para su uso por parte de las aplicaciones de la empresa en el inventario oficial de productos de terceros de la empresa y la licencia de código abierto asociada con el proyecto FOSS está presente en la licencia OSPO de sistema de inventario y está aprobado para la contribución.

3. Los proyectos (menores) de interés de la empresa no se pueden utilizar para publicar las contribuciones patentables o registradas de la empresa.

4. La solicitud de acceso al proyecto FOSS debe clasificarse como Interés (menor) de la empresa y debe cumplir los requisitos correspondientes a la categoría de Proyecto FOSS.

5. El gerente del solicitante debe aprobar la solicitud de acceso utilizando la herramienta aprobada por OSPO y debe asegurarse de que se cumplan los siguientes requisitos:

i. El producto FOSS está aprobado para su uso en las aplicaciones de la Compañía en el inventario oficial de productos de la Compañía.

ii. Es beneficioso para la Compañía permitir que el solicitante contribuya al proyecto (por ejemplo, el proyecto se utiliza en una aplicación comercial desarrollada por el solicitante).

iii. Si el proyecto requiere que los colaboradores acepten un Acuerdo de licencia de colaborador, un Certificado de origen del desarrollador o un acuerdo similar, se debe contactar al Departamento Legal a través de la OSPO para obtener asesoramiento.

iv. Realizar una revisión de propiedad intelectual e intereses en competencia para determinar la necesidad de una solicitud de patente o marca registrada antes de comenzar la contribución.

**Revisión y decisión de OSPO**

Las solicitudes de acceso a contribuciones a proyectos FOSS deben estar asociadas a una licencia aprobada en el inventario de licencias OSPO.

La OSPO tiene autoridad para tomar decisiones sobre las solicitudes de acceso y debe registrar su decisión en el sistema de registro si corresponde.

**Mantenimiento**

1. Las solicitudes de acceso al proyecto deben re-certificarse al menos una vez al año.

2. La licencia de código abierto del proyecto debe validarse regularmente para identificar, registrar y volver a evaluar los cambios de licencia de impacto.

**Transición/Retiro**

Los proyectos de contribución FOSS y cualquier solicitud de acceso asociada ya configurada en las herramientas aprobadas por OSPO deben ser retiradas dentro de los 3 meses posteriores a la retirada o el archivo de ese proyecto de su depósito público o del inventario oficial de productos de la empresa.

### Proyectos de contribución personal

Los empleados de la empresa pueden contribuir a los proyectos FOSS a título personal bajo las siguientes condiciones:

1. El proyecto FOSS no está registrado como proyecto moderado por la empresa o de interés de la empresa en el inventario FOSS.

2. La contribución al proyecto y el compromiso con la comunidad del proyecto no entra en conflicto con el código de conducta.

3. La contribución al proyecto y el compromiso con la comunidad del proyecto no deben representar un conflicto de intereses con los negocios de la Compañía.

4. La contribución al proyecto y el compromiso con la comunidad del proyecto no deben aprovecharse de la propiedad intelectual de la Compañía.

5. La contribución al proyecto y el compromiso con la comunidad del proyecto deben cumplir con el uso de pautas y gestión de publicaciones en redes sociales.

6. Los contribuidores no pueden presentarse a sí mismos como empleados de la Compañía.

7. Las contribuciones no deben hacerse desde hardware o infraestructura propiedad de la Compañía.

### Cumplimiento de las normas de exportación de tecnología

1. Los proyectos de interés para la empresa que estén moderados por ésta que proporcionan o realizan criptografía no estándar [1] deben revisarse antes de la aprobación de OSPO (consulte las [Regulaciones de exportación/importación de tecnología](https://github-com.translate.goog/Citi/citi-ospo/blob/main/FOSS%20Contribution%20Guidelines.md?_x_tr_sl=en&_x_tr_tl=es&_x_tr_hl=en&_x_tr_pto=wapp#technologyexportimportregulations)).

2. El personal de la empresa que participe en actividades de contribución FOSS debe cumplir con las normas de exportación locales.

## 3.7 Actividades de Contribución

1. Para todas las actividades de contribución que ocurran dentro de la red de la Compañía o en el hardware administrado por la Compañía, se deben implementar controles aprobados por OSPO para mitigar los riesgos descritos en Riesgo de contribución FOSS.

2. Las contribuciones FOSS realizadas desde el hardware administrado por la Compañía o ejecutadas en la red de la Compañía deben usar herramientas aprobadas por la Compañía.

3. Las contribuciones de FOSS realizadas desde hardware que no es de la empresa fuera de la red de la empresa no están restringidas al uso de herramientas.

### Proyectos moderados por la empresa

1. Se debe establecer un equipo de proyecto FOSS compuesto por las funciones definidas en "Funciones y responsabilidades".

2. El equipo del proyecto FOSS debe definir y documentar las instrucciones aprobadas por OSPO que describen cómo informar las vulnerabilidades de seguridad en el proyecto [2].

3. Los proyectos que ya no son moderados por un empleado de la Compañía deben pasar a Interés de la Compañía o Retirarse siguiendo el proceso descrito en la Sección "Transición/Retirada".

### Proyectos de interés (principal) para la empresa

1. Se debe establecer un equipo proyecto FOSS que consista en los roles correspondientes definidos en "Roles y responsabilidades".

2. Las actividades deben seguir los procesos de contribución establecidos por la comunidad del proyecto.

### Proyectos de (menor) interés para la empresa

Las actividades deben seguir procesos de contribución establecidos por la comunidad del proyecto.

### 5.1 Asistencia Jurídica

Las licencias gratuitas y de código abierto son acuerdos legales entre autores y consumidores. La OSPO mantiene un inventario de licencias FOSS aprobadas para la actividad de contribución por parte de los contribuyentes de la Compañía. El papel del departamento legal es asesorar a la OSPO. La OSPO se involucra legalmente según sea necesario al agregar una nueva licencia al inventario, cuando se producen cambios en una licencia existente, cuando los contribuyentes del proyecto acepten un Acuerdo de licencia de colaborador, Certificado de origen del desarrollador o acuerdo similar. Las actividades de revisión legal y la disposición serán capturadas y selladas por la OSPO.
