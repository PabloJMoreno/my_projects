<h4><center>Classification ANN (Multiclass Outcome)</center></h4>
<img src="https://i.ibb.co/5L43QPT/26.png" alt="Drawing" style="width: 500px;"/>

```python
data2 = data[['mileage','year']]
data2.head()
```

<h5><center>Weights</center></h5>
<img src="https://i.ibb.co/z6ssKpt/27.png" alt="Drawing" style="width: 500px;"/>

```python
data2['weight1'] = .0000743856426
data2['weight2'] = .0000291011971
data2['weight3'] = -.000132187805
data2['weight4'] = -.00373746426
data2['weight5'] = -.000789461808
data2['weight6'] = .00256261417
data2.head()
```

<h5><center>Biases</center></h5>
<img src="https://i.ibb.co/gDxpmwQ/28.png" alt="Drawing" style="width: 500px;"/>

```python
data2['bias1'] = -.00000185693427
data2['bias2'] = -.000000392653538
data2['bias3'] = .00000127350364

data2.head()
```

<h5><center>Linear Equation</center></h5>
<img src="https://i.ibb.co/HVMnSzD/29.png" alt="Drawing" style="width: 500px;"/>

```python
data2['linear1'] = data2['mileage'] * data2['weight1'] + data2['year'] * data2['weight4'] + data2['bias1']
data2['linear2'] = data2['mileage'] * data2['weight2'] + data2['year'] * data2['weight5'] + data2['bias2']
data2['linear3'] = data2['mileage'] * data2['weight3'] + data2['year'] * data2['weight6'] + data2['bias3']
data2.head()
```

<h5><center>Softmax Activation Function</center></h5>
<img src="https://i.ibb.co/whNY4my/31.png" alt="Drawing" style="width: 500px;"/>

```python
# Activation Function
import numpy as np
data2['prob_sold']     = np.exp(data2['linear1']) / (np.exp(data2['linear1']) + np.exp(data2['linear3']) + np.exp(data2['linear2']))
data2['prob_not_sold'] = np.exp(data2['linear2']) / (np.exp(data2['linear1']) + np.exp(data2['linear3']) + np.exp(data2['linear2']))
data2['prob_leased']   = np.exp(data2['linear3']) / (np.exp(data2['linear1']) + np.exp(data2['linear3']) + np.exp(data2['linear2']))

data2.head()
```

<h5><center>Prediction</center></h5>
<img src="https://i.ibb.co/DVm4x2d/32.png" alt="Drawing" style="width: 500px;"/>

```python
data2['prediction'] = np.where((data2['prob_sold'] >= data2['prob_not_sold']) & (data2['prob_sold'] >= data2['prob_leased']),
                               1, np.where((data2['prob_not_sold'] >= data2['prob_sold']) & (data2['prob_not_sold'] >= data2['prob_leased']),
                               2, 3))           
data2.head()
```

```python
# Accuracy
accuracy_score(data['sold'], data2['prediction'])
```
