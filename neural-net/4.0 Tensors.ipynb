```python
import PIL.Image as Image

sunset = tf.keras.utils.get_file('640px-Sunset-Sicilia-bjs-1.jpg','https://upload.wikimedia.org/wikipedia/commons/thumb/c/cc/Sunset-Sicilia-bjs-1.jpg/640px-Sunset-Sicilia-bjs-1.jpg')
sunset = Image.open(sunset)
sunset
```

```python
# There can be an arbitrary number of
# axes (sometimes called "dimensions")
rank_3_tensor = tf.constant(sunset)
print(rank_3_tensor)
```
